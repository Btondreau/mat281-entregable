4. **N칰mero de t칤tulos agregados por a침o**

   * Pregunta gu칤a: 쮺u치ndo creci칩 m치s el cat치logo de Netflix? 쯈u칠 relaci칩n podr칤a tener con la expansi칩n internacional de la plataforma?

6. **Duraci칩n de pel칤culas por g칠nero**

   * Extrae la duraci칩n en minutos y crea un boxplot agrupado por g칠nero principal.
   * Pregunta gu칤a: 쯈u칠 g칠neros suelen durar m치s? 쯈u칠 podr칤as decir sobre la variabilidad dentro de un mismo g칠nero?


### 4

```python
# Asegurar que la columna sea de tipo fecha
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Crear columna con el a침o de agregado
df['year_added'] = df['date_added'].dt.year

# Contar cu치ntos t칤tulos se agregaron por a침o
titulos_por_a침o = df['year_added'].value_counts().sort_index()

# Gr치fico de l칤nea (serie temporal)
plt.figure(figsize=(12,6))
sns.lineplot(x=titulos_por_a침o.index, y=titulos_por_a침o.values, marker="o")
plt.title("N칰mero de t칤tulos agregados al cat치logo de Netflix por a침o")
plt.xlabel("A침o en que se agreg칩")
plt.ylabel("Cantidad de t칤tulos")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
print("\nEl c치talogo de Netflix creci칩 m치s entre el a침o 2016 y 2017, pero el aumento empezo desde 2015 en adelante.")
print("Tiene una relaci칩n directa con la expansi칩n internacional de la plataforma, ya que aproximadamente en esos a침os se empez칩 a masificar.")
```

### 5

```python
# 5. Heatmap de lanzamientos por a침o y mes

# Crear columnas de a침o y mes de agregado
df['month_added'] = df['date_added'].dt.month

# Eliminar filas con fechas nulas si las hay
df_cleaned = df.dropna(subset=['year_added', 'month_added'])

# Crear tabla pivote para el heatmap
heatmap_data = df_cleaned.pivot_table(index='year_added', columns='month_added', aggfunc='size', fill_value=0)

# Mapear n칰meros de mes a nombres
month_names = {1: 'Ene', 2: 'Feb', 3: 'Mar', 4: 'Abr', 5: 'May', 6: 'Jun',
               7: 'Jul', 8: 'Ago', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dic'}
heatmap_data.columns = [month_names[col] for col in heatmap_data.columns]


# Crear el heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='YlGnBu', linewidths=.5)
plt.title('N칰mero de T칤tulos Agregados por A침o y Mes')
plt.xlabel('Mes')
plt.ylabel('A침o')
plt.show()
print("\nS칤, hay meses con m치s estrenos que otros, en particular en los meses del segundo semestre.")
print("Las conclusiones que puedo sacar sobre la estacionalidad de lanzamientos es que coicide con que en segundo semestre existen mayores festividades, como por ejemplo las fiestas de fin de a침o y Halloween, a la vez de que en el hemisferio norte se encuentran en verano, que es cuando la demanda de entretenimiento aumenta.")
```

### 6

```python
# Filtrar solo pel칤culas
movies = df[df['type'] == 'Movie'].copy()

# Extraer duraci칩n en minutos
movies['duration_min'] = movies['duration'].str.replace(" min", "", regex=False)
movies['duration_min'] = pd.to_numeric(movies['duration_min'], errors='coerce')

# Extraer g칠nero principal (primer valor de listed_in)
movies['genre_main'] = movies['listed_in'].str.split(',').str[0].str.strip()

# Boxplot por g칠nero
plt.figure(figsize=(14,6))
sns.boxplot(data=movies, x='genre_main', y='duration_min')
plt.xticks(rotation=90)
plt.title("Distribuci칩n de la duraci칩n de pel칤culas por g칠nero")
plt.xlabel("G칠nero principal")
plt.ylabel("Duraci칩n (minutos)")
plt.show()
print("\nLos g칠neros que suelen durar m치s son: Dramas, Action & Adventure y Classic Movies.")
print("La variabilidad es amplia en algunos g칠neros, pero mucho menor en otros lo que refleja la flexibilidad o estandarizaci칩n propia de cada categor칤a, donde a cada una le da una identidad.")
```

```python
# Separar pa칤ses y normalizar
countries = df['country'].dropna().str.split(', ')
countries_exploded = countries.explode()

# Contar frecuencias de pa칤ses
top_countries = countries_exploded.value_counts().head(10)

# Gr치fico de barras
plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette="viridis")
plt.title("Top 10 pa칤ses con m치s producciones en Netflix")
plt.xlabel("Cantidad de t칤tulos")
plt.ylabel("Pa칤s")
plt.show()
print("\nLos pa칤ses que dominan el c치talogo son: United Stetes (muy por encima de los dem치s), India y United Kingdom.")
print("La distribuci칩n es esperada, ya que Netflix es una empresa estadounidense , hecha y creada para ellos, adem치s de que es el pa칤s que genera las producciones m치s comerciales. India no sorpende, porque es el pa칤s despu칠s de United States que m치s craci칩n de producciones tiene en el mundo. Y United Kingdom no sorprende, puesto que su lengua nativa tambi칠n es el ingl칠s.")
```

### 8

```python
# Extraer g칠nero principal
df['genre_main'] = df['listed_in'].str.split(',').str[0].str.strip()

# Contar t칤tulos por g칠nero y tipo
conteo_genero_tipo = df.groupby(['genre_main', 'type']).size().unstack(fill_value=0)

# Seleccionar los g칠neros m치s frecuentes (ej. top 15 para que no se vea saturado)
top_genres = conteo_genero_tipo.sum(axis=1).sort_values(ascending=False).head(15)
conteo_top = conteo_genero_tipo.loc[top_genres.index]

# Gr치fico de barras apiladas
conteo_top.plot(kind='bar', stacked=True, figsize=(12,6), color=['red','blue'])
plt.title("Pel칤culas vs Series seg칰n g칠nero")
plt.xlabel("G칠nero principal")
plt.ylabel("Cantidad de t칤tulos")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Tipo")
plt.show()
print("\nS칤, existen g칠neros m치s asociados a pel칤culas o series. Por ejemplo, Dramas y Comedies est치n m치s asociado a pel칤culas, en cambio, International TV Shows yCrime TV Shows est치n m치s asociado a series.")
```

### 9

```python
# Filtrar solo pel칤culas
movies = df[df['type'] == 'Movie'].copy()

# Extraer duraci칩n en minutos
movies['duration_min'] = movies['duration'].str.replace(" min", "", regex=False)
movies['duration_min'] = pd.to_numeric(movies['duration_min'], errors='coerce')

# Gr치fico de dispersi칩n
plt.figure(figsize=(12,6))
sns.scatterplot(data=movies, x='release_year', y='duration_min', alpha=0.4)
plt.title("Duraci칩n de pel칤culas vs A침o de lanzamiento")
plt.xlabel("A침o de lanzamiento")
plt.ylabel("Duraci칩n (minutos)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
print("No sse aprecia que hayan cambiado los est치ndares de duraci칩n de las pel칤culas con el tiempo.")
print("Si se ven outliers interesantes. Se ven pel칤culas extremadamente largas y otras muy cortas, pero que es menor en comparaci칩n al otro extremo. Esto lo que refleja es la diversidad de la oferta de Netflix.")
```
