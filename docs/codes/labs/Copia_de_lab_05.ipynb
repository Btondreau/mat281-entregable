```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
```

```python
# Cargar datos
df = pd.read_csv('https://raw.githubusercontent.com/fralfaro/MAT281/main/docs/labs/data/netflix_titles.csv')
df.head()
```

### 1.

```python
# Contar cantidad de películas y series
conteo_tipos = df['type'].value_counts()

# Gráfico con matplotlib
plt.figure(figsize=(6,4))
conteo_tipos.plot(kind='bar', color=['red', 'blue'])
plt.title("Distribución de tipos de contenido en Netflix")
plt.xlabel("Tipo de contenido")
plt.ylabel("Cantidad de títulos")
plt.xticks(rotation=0)
plt.show()
print("\nEl tipo de contenido predominante en Netflix es Movie.")
print("La implicancia que podría tener para los usarios es que tiene más opciones de movie que elegir TV Show, así teniendo más usuarios de consumo a corto plazo que a largo.")
```

### 2.

```python
# Histograma con matplotlib
plt.figure(figsize=(10,6))
plt.hist(df['release_year'], bins=30, color='purple', edgecolor='black')
plt.title("Distribución de títulos por año de lanzamiento")
plt.xlabel("Año de lanzamiento")
plt.ylabel("Cantidad de títulos")
plt.show()
print("\nHya más contenido reciente, ya que la cantidad de títulos son principlamente después 2000.")
print("Se nota un patrón en las décadas anteriores al 2000, donde es una frecuencia de títulos muy baja, a la vez que se nota un patrón en la década del 2010, el cual consiste en un crecimiento constante en la cantidad de títulos.")
```

### 3

```python
# Conteo de clasificaciones por edad
conteo_rating = df['rating'].value_counts()

# Gráfico de barras
plt.figure(figsize=(10,6))
sns.barplot(x=conteo_rating.index, y=conteo_rating.values, palette="pastel")
plt.title("Distribución de clasificaciones por edad en Netflix")
plt.xlabel("Clasificación por edad")
plt.ylabel("Cantidad de títulos")
plt.xticks(rotation=45)
plt.show()
print("\nLa clasificación más común es TV-MA")
print("Creo que Netflix está más enfocado público adulto y adolcentes.")
```



## 🎨 Parte 2: Tendencias y evolución en el tiempo

4. **Número de títulos agregados por año**

   * Usa `date_added` (convertido a fecha) y grafica una serie temporal.
   * Pregunta guía: ¿Cuándo creció más el catálogo de Netflix? ¿Qué relación podría tener con la expansión internacional de la plataforma?

5. **Heatmap de lanzamientos por año y mes**

   * Construye un mapa de calor con `year_added` y `month_added`.
   * Pregunta guía: ¿Hay meses con más estrenos que otros? ¿Qué conclusiones puedes sacar sobre la estacionalidad de lanzamientos?

6. **Duración de películas por género**

   * Extrae la duración en minutos y crea un boxplot agrupado por género principal.
   * Pregunta guía: ¿Qué géneros suelen durar más? ¿Qué podrías decir sobre la variabilidad dentro de un mismo género?


### 4

```python
# Asegurar que la columna sea de tipo fecha
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Crear columna con el año de agregado
df['year_added'] = df['date_added'].dt.year

# Contar cuántos títulos se agregaron por año
titulos_por_año = df['year_added'].value_counts().sort_index()

# Gráfico de línea (serie temporal)
plt.figure(figsize=(12,6))
sns.lineplot(x=titulos_por_año.index, y=titulos_por_año.values, marker="o")
plt.title("Número de títulos agregados al catálogo de Netflix por año")
plt.xlabel("Año en que se agregó")
plt.ylabel("Cantidad de títulos")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
print("\nEl cátalogo de Netflix creció más entre el año 2016 y 2017, pero el aumento empezo desde 2015 en adelante.")
print("Tiene una relación directa con la expansión internacional de la plataforma, ya que aproximadamente en esos años se empezó a masificar.")
```

### 5

```python
# 5. Heatmap de lanzamientos por año y mes

# Crear columnas de año y mes de agregado
df['month_added'] = df['date_added'].dt.month

# Eliminar filas con fechas nulas si las hay
df_cleaned = df.dropna(subset=['year_added', 'month_added'])

# Crear tabla pivote para el heatmap
heatmap_data = df_cleaned.pivot_table(index='year_added', columns='month_added', aggfunc='size', fill_value=0)

# Mapear números de mes a nombres
month_names = {1: 'Ene', 2: 'Feb', 3: 'Mar', 4: 'Abr', 5: 'May', 6: 'Jun',
               7: 'Jul', 8: 'Ago', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dic'}
heatmap_data.columns = [month_names[col] for col in heatmap_data.columns]


# Crear el heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='YlGnBu', linewidths=.5)
plt.title('Número de Títulos Agregados por Año y Mes')
plt.xlabel('Mes')
plt.ylabel('Año')
plt.show()
print("\nSí, hay meses con más estrenos que otros, en particular en los meses del segundo semestre.")
print("Las conclusiones que puedo sacar sobre la estacionalidad de lanzamientos es que coicide con que en segundo semestre existen mayores festividades, como por ejemplo las fiestas de fin de año y Halloween, a la vez de que en el hemisferio norte se encuentran en verano, que es cuando la demanda de entretenimiento aumenta.")
```

### 6

```python
# Filtrar solo películas
movies = df[df['type'] == 'Movie'].copy()

# Extraer duración en minutos
movies['duration_min'] = movies['duration'].str.replace(" min", "", regex=False)
movies['duration_min'] = pd.to_numeric(movies['duration_min'], errors='coerce')

# Extraer género principal (primer valor de listed_in)
movies['genre_main'] = movies['listed_in'].str.split(',').str[0].str.strip()

# Boxplot por género
plt.figure(figsize=(14,6))
sns.boxplot(data=movies, x='genre_main', y='duration_min')
plt.xticks(rotation=90)
plt.title("Distribución de la duración de películas por género")
plt.xlabel("Género principal")
plt.ylabel("Duración (minutos)")
plt.show()
print("\nLos géneros que suelen durar más son: Dramas, Action & Adventure y Classic Movies.")
print("La variabilidad es amplia en algunos géneros, pero mucho menor en otros lo que refleja la flexibilidad o estandarización propia de cada categoría, donde a cada una le da una identidad.")
```

### 7

```python
# Separar países y normalizar
countries = df['country'].dropna().str.split(', ')
countries_exploded = countries.explode()

# Contar frecuencias de países
top_countries = countries_exploded.value_counts().head(10)

# Gráfico de barras
plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette="viridis")
plt.title("Top 10 países con más producciones en Netflix")
plt.xlabel("Cantidad de títulos")
plt.ylabel("País")
plt.show()
print("\nLos países que dominan el cátalogo son: United Stetes (muy por encima de los demás), India y United Kingdom.")
print("La distribución es esperada, ya que Netflix es una empresa estadounidense , hecha y creada para ellos, además de que es el país que genera las producciones más comerciales. India no sorpende, porque es el país después de United States que más cración de producciones tiene en el mundo. Y United Kingdom no sorprende, puesto que su lengua nativa también es el inglés.")
```

### 8

```python
# Extraer género principal
df['genre_main'] = df['listed_in'].str.split(',').str[0].str.strip()

# Contar títulos por género y tipo
conteo_genero_tipo = df.groupby(['genre_main', 'type']).size().unstack(fill_value=0)

# Seleccionar los géneros más frecuentes (ej. top 15 para que no se vea saturado)
top_genres = conteo_genero_tipo.sum(axis=1).sort_values(ascending=False).head(15)
conteo_top = conteo_genero_tipo.loc[top_genres.index]

# Gráfico de barras apiladas
conteo_top.plot(kind='bar', stacked=True, figsize=(12,6), color=['red','blue'])
plt.title("Películas vs Series según género")
plt.xlabel("Género principal")
plt.ylabel("Cantidad de títulos")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Tipo")
plt.show()
print("\nSí, existen géneros más asociados a películas o series. Por ejemplo, Dramas y Comedies están más asociado a películas, en cambio, International TV Shows yCrime TV Shows están más asociado a series.")
```

### 9

```python
# Filtrar solo películas
movies = df[df['type'] == 'Movie'].copy()

# Extraer duración en minutos
movies['duration_min'] = movies['duration'].str.replace(" min", "", regex=False)
movies['duration_min'] = pd.to_numeric(movies['duration_min'], errors='coerce')

# Gráfico de dispersión
plt.figure(figsize=(12,6))
sns.scatterplot(data=movies, x='release_year', y='duration_min', alpha=0.4)
plt.title("Duración de películas vs Año de lanzamiento")
plt.xlabel("Año de lanzamiento")
plt.ylabel("Duración (minutos)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
print("No sse aprecia que hayan cambiado los estándares de duración de las películas con el tiempo.")
print("Si se ven outliers interesantes. Se ven películas extremadamente largas y otras muy cortas, pero que es menor en comparación al otro extremo. Esto lo que refleja es la diversidad de la oferta de Netflix.")
```
